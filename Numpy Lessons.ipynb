Agora, vejamos um primeiro exemplo de criação de um array Numpy:

```python
a = np.array([0, 1, 2, 3])

print(a)
print(type(a))
print(a.ndim)
```

## Criação de Arrays

### Construção Manual de Arrays

```python
b = np.array([[0, 1, 2], [3, 4, 5]])

print(b)
print(b.ndim)
print(b.shape)
print(len(b))
```

### Criação utilizando funções

```python
c = np.arange(10, 20)

print(c)

d = np.linspace(1, 10, 13, endpoint=False)
print(d)
```

### Arrays comuns

```python
e = np.zeros(5)

print(e)

f = np.random.rand(10, 10)
print(f)
```

## Indexação e Cortes (Slices) de Arrays

```python
a = np.arange(10, 21)

print(a)

print(a[2:9])
print(a[8:2:-1])
```

```python
b = np.random.rand(5, 5)

print(b)

print('\n')

print(b[0:2, 2])
```
