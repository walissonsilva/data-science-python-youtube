# #15 - Método Apply e Ordenação de DataFrames
---

- Médoto `apply`
- Método `sort_values`

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/titanic_train.csv')
```

```python
df.head()
```

```python
df['Age'].apply(np.sqrt)
```

```python
def changeSurvived(elemento):
    if elemento == 0:
        return 'No'
    else:
        return 'Yes'

df['Survived'] = df['Survived'].apply(changeSurvived)
```

```python
df.head()
```

```python
df['Fare'].dtype
```

```python
def changeFare(elemento):
    return '${:.2f}'.format(elemento)

df['Fare'] = df['Fare'].apply(changeFare)
```

```python
df['Fare']
```

## Ordenação

```python
df_sort = df.sort_values(by='Age')
```

```python
df_sort
```

```python
df_sort.dropna()
```

```python
df_sort = df.sort_values(by='Age', ascending=False, kind='mergesort')
```

```python
df_sort
```

```python
df_sort.dropna()
```
