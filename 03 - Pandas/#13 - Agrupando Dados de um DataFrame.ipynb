# #13 - Agrupando Dados de um DataFrame

Método `groupby`: [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/groupby.html)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/pesquisa-covid.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df_covid = df.groupby('COVID-19')
```

```python
df_covid
```

```python
df_covid.mean()
```

```python
df_covid.sum()
```
